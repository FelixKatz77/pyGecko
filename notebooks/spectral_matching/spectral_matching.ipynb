# Spectral matching with pyGecko

This notebook gives an example on how to perform spectral comparisons with pyGecko. Our goal is to identify a product in a GC-MS chromatogram based on the spectrum of a reference sample.

```python
from pygecko import MS_Base_Parser
from pygecko.gc_tools import Spectral_Match
from pygecko.visualization import Visualization
```

First we load the injection with the reference sample and pick peaks.

```python
reference_injection = MS_Base_Parser.load_injection('FKB-FA-062-Prod.mzML')
reference_injection.pick_peaks()
print(reference_injection)
```

Next we repeat the same steps for the injection with the sample potentially containing the product.

```python
injection_d1 = MS_Base_Parser.load_injection('FBS-FA-034-D1.mzML')
injection_d1.pick_peaks()
print(injection_d1)
```

We can access the peak with our reference product by indexing the injection with the retention time of the peak.

```python
reference_peak = reference_injection[9.409]
```

To find the peak in the injection that matches the reference we use the find_peak method of the Spectral_Match class.

```python
match = Spectral_Match.find_peak(reference_peak, injection_d1)
```

We can visualize the match by plotting the spectra of the reference and the matched peak.

```python
Visualization.compare_mass_spectra(match)
```
